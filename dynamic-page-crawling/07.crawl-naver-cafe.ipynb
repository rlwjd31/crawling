# 네이버 카페 크롤링

- 유투브 후기 이동
- contents를 50개식 출력하게 하는 filter 부분 클릭하기

추출 데이터
- 제목
- 링크

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from urllib.parse import urljoin

ROOT_URL = "https://cafe.naver.com/startcodingofficial"
PATH = "/startcodingofficial"

driver = webdriver.Chrome()
driver.get(urljoin(ROOT_URL, PATH))
```

```python
# 유투부 후기 링크 클릭
driver.find_element(By.CSS_SELECTOR, "#menuLink15").click()
```

```python
select_box_selector = ".sort_area .FormSelectBox .FormSelectButton"
driver.find_element(By.CSS_SELECTOR, select_box_selector).click()
# 50개씩 선택
driver.find_element(By.CSS_SELECTOR, "div.select_option > ul > li:nth-child(7) > button").click()
```

```python
articles = driver.find_elements(By.CSS_SELECTOR, ".article-board > table > tbody:nth-child(5) > tr")

# print(articles)
results = {
  "제목": [],
  "링크": []
}
for article in articles:
  results["제목"].append(article.find_element(By.CSS_SELECTOR, ".board-list a").text)
  results["링크"].append(article.find_element(By.CSS_SELECTOR, ".board-list a").get_attribute("href"))
```

```python
import pandas as pd

df = pd.DataFrame(results)
df.to_excel("./outputs/naver_cafe_youtube_review.xlsx", index=False)
```
